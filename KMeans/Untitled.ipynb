```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
```

```python
x,y=make_blobs(n_samples=500,
    n_features=2,
    centers=5,
    shuffle=True,
    random_state=3)
```

```python
print(x)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.grid("on")
plt.show()
```

```python
k=5
colors=["red","orange","blue","violet","green"]
```

```python
clusters={}
```

```python
for i in range(k):
    center=20*np.random.random((x.shape[1],))-10
    points=[]
    cluster={
        'center':center,
        'point':points,
        'color':colors[i]
         }
    clusters[i]=cluster
```

```python
print(clusters)
```

```python
def dista(p,q):
    return np.sqrt(np.sum((p-q)**2))
```

```python
for ix in range(x.shape[0]):
    dist=[]
    curr_x=x[ix]
    for i in range(k):
        d=dista(curr_x,clusters[i]['center'])
        dist.append(d)
    cluster_c=np.argmin(dist)
    clusters[cluster_c]['point'].append(curr_x)
```
