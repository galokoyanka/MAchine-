```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("download.jpg")
```

```python
print(img)
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
img.shape
```

```python
X=img.reshape((-1,3))
```

```python
X.shape
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=3)
```

```python
kmeans.fit(X)
```

```python
centers=kmeans.cluster_centers_
```

```python
print(centers)
```

```python
import numpy as np

colors=[]
for i,colors in enumerate(centers):
    plt.subplot(1,3,i+1)
    square=np.zeros((100,100,3),dtype='uint8')
    square[:,:,:]=colors
    plt.imshow(square)
    plt.axis("off")
```
