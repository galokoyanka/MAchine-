```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
a=np.array([1,2,3,4,5,6,7])
a=a**2
a1=a+2
print(a)
print(a1)
```

```python
plt.style.available
```

```python
plt.style.use("dark_background")
plt.figure(figsize=(20,20))
plt.plot(a)
plt.plot(a1,color="black",label="stock price")
plt.legend()
plt.xlabel("days")
plt.ylabel("stock")
plt.plot(a,a1)
plt.scatter(a,a1,color="green")
plt.show()
```

# BAr chart

```python
import numpy as ag
x=ag.array([1,2,3,4])
y=ag.array([1,4,9,16])
z=ag.array([1,8,27,64])
plt.bar(x,z,0.2,color='green',label="cubes")
plt.bar(x+0.2,y,0.2,color='red',label="squares",tick_label=["one","TWo","Three","Four"])

plt.legend()
plt.show()
```
