```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dfx=pd.read_csv("Linear_X_Train.csv").values
dfy=pd.read_csv("Linear_Y_Train.csv").values
dft=pd.read_csv("Linear_X_Test.csv").values
```

```python
print(dfx)
print(dfx.shape)
print(dfy)
print(dfy.shape)
```

```python
plt.scatter(dfx,dfy)
plt.show()
```

```python
def grad(X,Y,theta):
    
    m = X.shape[0]
    grad = np.zeros((2,))
    
    for i in range(m):
        yi = theta[1]*X[i] + theta[0]
        grad[0] += 2*(yi - Y[i])
        grad[1] += 2*(yi - Y[i])*X[i]
    
    return grad
```

```python
def gradientdecent(X,Y,learning_rate=.0001,max_steps=1000):
    Theta=np.zeros((2,))      
    for i in range(max_steps):
        g=grad(X,Y,Theta)
        Theta[0] = Theta[0]-learning_rate*g[0]
        Theta[1] = Theta[1]-learning_rate*g[1]
              
    return Theta
        
```

```python
theta = gradientdecent(dfx[:1000],dfy[:1000])
```

```python
print(theta)
```

```python
def predict(x_test,theta):
    return theta[0]+theta[1]*x_test
```

```python
x_test=dft
```

```python
print(x_test)
Y_pred=predict(x_test,theta)
print(Y_pred)
```

```python
plt.scatter(dfx,dfy)
plt.plot(x_test,Y_pred,color="orange")
plt.show()
```

```python
np.savetxt("f20.csv", Y_pred, delimiter=",")
```

```python
lr.fit(dfx,dfy)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
lr.score(dfx,dfy)
```

```python
def predict(x_test,x,y):
    return y+x*x_test
```

```python
Y_pred=predict(x_test,lr.coef_,lr.intercept_)
print(Y_pred)
```

```python
lr.score(x_test,Y_pred)
```
